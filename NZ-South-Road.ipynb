### Import Libraries and Unzip files for Road network data

```python
import zipfile
import os
import geopandas as gpd
import matplotlib.pyplot as plt

# Define the path to the zip file
zip_file_path = "road_network.zip"

# Get the directory where the zip file is located
extract_dir = os.path.dirname(zip_file_path)

# Unzip the file into the same directory
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)
```

### Plot Road Network

```python
import zipfile
import os
import geopandas as gpd
import matplotlib.pyplot as plt

# Get the path of the .shp file (adjust the file name if needed)
shapefile_path = os.path.join(extract_dir, "road_network.shp")

# Load the shapefile using GeoPandas
gdf = gpd.read_file(shapefile_path)

# Plot the road network
gdf.plot()

# Show the plot
plt.show()
```

### Check the Coordinate Reference System (CRS)

```python
print("Coordinate Reference System (CRS):")
print(gdf.crs)
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Path to your local shapefile
shapefile_path = "road_network.shp"

# Load the shapefile using GeoPandas
gdf = gpd.read_file(shapefile_path)

# Plot the road network with a title and labels
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, legend=True)

# Add title and labels
ax.set_title('Road Network Map', fontsize=16)
ax.set_xlabel('Longitude', fontsize=12)
ax.set_ylabel('Latitude', fontsize=12)

# Show the plot
plt.show()
```
